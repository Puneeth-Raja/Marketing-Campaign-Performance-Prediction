```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

# Select only your numeric columns
num_cols = ['Duration', 'Impressions', 'Clicks', 'Leads', 'Conversions', 'Revenue', 'Acquisition_Cost', 'ROI', 'Engagement_Score']

# Initialize the imputer (it uses Bayesian Ridge regression by default)
imputer = IterativeImputer(random_state=42, max_iter=10)

# Fit and transform the numeric data
df_all[num_cols] = imputer.fit_transform(df_all[num_cols])
```

```python
# Commit to the clean, dropped dataset
df_all = df_all.dropna()

# Step 2: Fix the low-end negative artifacts created by the IterativeImputer
zero_floor_cols = ['Impressions', 'Clicks', 'Leads', 'Conversions', 'Revenue', 'Acquisition_Cost', 'Engagement_Score']
for col in zero_floor_cols:
    df_all[col] = df_all[col].clip(lower=0)

# ROI can legally be negative (loss), but clamp any weird imputer math below -1.0 (100% loss)
df_all['ROI'] = df_all['ROI'].clip(lower=-1.0)

# Step 3: Winsorize the extreme high-end spikes at the 99th percentile
skewed_cols = ['Revenue', 'Acquisition_Cost', 'ROI', 'Clicks', 'Leads', 'Conversions']
for col in skewed_cols:
    q_upper = df_all[col].quantile(0.99)
    df_all[col] = df_all[col].clip(upper=q_upper)

print("Preprocessing complete! Final clean dataset shape:", df_all.shape)
```

```python
df_all.describe()
```

```python
# 1. Calculate Total Acquisition Cost
df_all['Total_Acquisition_Cost'] = df_all['Conversions'] * df_all['Acquisition_Cost']

# 2. Calculate Net Profit
df_all['Net_Profit'] = df_all['Revenue'] - df_all['Total_Acquisition_Cost']

# 3. Calculate True ROI (Adding 1e-5 to avoid division by zero)
df_all['ROI'] = df_all['Net_Profit'] / (df_all['Total_Acquisition_Cost'] + 1e-5)

# 4. Create Profit_Loss Target Column (1 for Profit, 0 for Loss)
df_all['Profit_Loss'] = (df_all['Net_Profit'] > 0).astype(int)
```
